## 1. Conversion d'une base b vers la base décimale

### 1.1 Méthode de droite à gauche

Voici un algorihtme permettant de convertir un nombre entier écrit en base $b$ vers la base décimale :
> resultat $\leftarrow$ 0<br>
> i $\leftarrow$ 0<br>
> **Pour** tous les chiffres $c$ allant de droite à gauche : <br>
>>    resultat $\leftarrow$ resultat + $c \times b^i$<br>
>>    i $\leftarrow$ i+1<br>
>
> Renvoyer resultat<br>

```python
# Question 1

def convert_dec(toConvert, base) :
    """Prend en entrée un nombre entier à convertir en base décimale et la base actuelle de ce nombre et affiche le
    nombre converti en base décimale comme résultat. Syntaxe : convert_dec(nbEntree, base)"""
    resultat = 0
    paramStr = str(toConvert)  #Crée une variante string de la variable d'entrée pour pouvoir récupérer des outs de celle-ci
    paramLengh = len(paramStr) #Récupère la taille de la variable entrée pour la boucle for
    
    for i in range(paramLengh) :
        resultat += int(paramStr[i])*base**(paramLengh-i-1) #Calcul long et compliqué
        
    print(paramStr+" en base "+str(base)+" vaut "+str(resultat)+" en base 10") #écrit joliment le résultat

convert_dec(325, 8)
```

### 1.2 Méthode de gauche à droite (méthode de Horner)

Voici un 2ème algorihtme permettant aussi de convertir un nombre entier écrit en base $b$ vers la base décimale mais en parcourant le nombre dans l'autre sens :
> resultat $\leftarrow$ 0<br>
> **Pour** tous les chiffres $c$ allant de gauche à droite : <br>
>>    resultat $\leftarrow$ b $\times$ resultat<br>
>>    resultat $\leftarrow$ resultat + $c$<br>
>
> Renvoyer resultat<br>

**Remarque :** Cette méthode peut paraître plus complexe que la première mais elle est en fait bien plus efficace (rapide) car elle ne nécessite que des opérations simples (multiplication et addition) au lieu d'utiliser des puissances.

## 2. Conversion la base décimale vers une base b

### 2.1 Conversion par divisions successives par $b$
Pour convertir un nombre entier de la base décimale vers une base $b$, il suffit de le diviser par $b$ (division entière), de noter le reste puis de diviser à nouveau le résultat par $b$ et de noter le reste et ainsi de suite jusqu'à obtenir 0. En écrivant les restes obtenus en commençant par le dernier, on obtient le résultat de la conversion.

**Q5. Programmer une fonction ``convert_base`` qui, selon l'algorithme précédent, prend en argument un nombre entier dans la base décimale et la base dans laquelle on veut le convertir et renvoie la valeur convertie. Penser à spécifier la fonction et éventuellement à utiliser ``assert``.**

```python
# Question 5

def convert_base(toConvert, base) :
    """ Prend un entier en base 10 et un entier correspondant à la base souhaité en entrée et affiche l'entier entré
    traduit en la base souhaité comme résultat. Syntaxe : convert_base(entierAConvertir, baseSouhaité)"""
    resultat = ""
    finalResultat = ""
    paramStr = str(toConvert)  #Crée une variante string de la variable d'entrée pour pouvoir récupérer des outs de celle-ci
    paramLengh = len(paramStr) #Récupère la taille de la variable entrée pour la boucle for
    
    for i in range(paramLengh) :
        resultat += str(toConvert%base) #Calcul
        toConvert = toConvert//base     #Calcul
    
    for i in range(len(resultat)) :
        finalResultat += resultat[len(resultat)-i-1] #inverse le résultat (lit à l'envers)
    
    print(paramStr+" en base 10 vaut "+finalResultat+" en base "+str(base)) #écrit joliment le résultat
        
convert_base(213, 8)
```

### 2.2 Conversion par divisions successives par $b^i$
On utilisera plutôt cette méthode si on veut fixer le nombre $k$ de chiffres utilisés pour écrire le résultat, quitte à ce que ce résultat commence par des 0.

Il suffit alors d'effectuer la division entière du nombre à convertir par $b^{k-1}$ puis de recommencer en divisant le reste par $b^{k-2}$ et ainsi de suite judsqu'à diviser par $b^0=1$. Les résultats successifs de ces divisions dans l'ordre forment le nombre final. Evidemment, cela ne fonctionne pas si le nombre à convertir est trop grand pour le nombre de chiffres choisi pour le résultat.

Dans le cas d'une conversion vers le binaire, l'agorithme est le suivant :
> **Pour** $i$ allant de $k-1$ à $0$ : <br>
>> **Si** $n \geq b^i$ :<br>
>>>    $n \leftarrow n-b^i $<br>
>>>    Rajouter le chiffre $1$ au résultat<br>
>>
>> **Sinon** :<br>
>>> Rajouter le chiffre $0$ au résultat<br>
>
> Renvoyer resultat<br>

**Q7. Programmer une fonction ``convert_bin`` qui, selon l'algorithme précédent, prend en argument un nombre entier dans la base décimale et le nombre de chiffres souhaité pour le résultat et renvoie la valeur convertie en binaire. Penser à spécifier la fonction et éventuellement à utiliser ``assert``.**

```python
# Question 7

def convert_bin(n, k) :
    """ Prend un entier en base décimale et le nombre de chiffres souhaités pour le résultat en entrée et affiche la
    valeur d'entrée convertie en binaire comme résultat. Syntaxe : convert_bin(nbAConvertir, longueurResultat) """
    resultat = ""
    for i in range(k-1,-1,-1) : #calculs
        if (n >= 2**i) :
            n -= 2**i
            resultat += "1"
        else :
            resultat += "0"
    
    print(resultat) #écrit le résultat
    
convert_bin(131, 8)
```
