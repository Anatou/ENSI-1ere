## 1. Calculer une distance

En Math√©matiques la distance entre deux points $A(x_A,y_A)$ et $B(x_B,y_B)$ dont on conna√Æt les coordonn√©es dans un rep√®re orthonorm√© est donn√©e par la relation : $$AB = \sqrt{(x_B - x_A)^2 + (y_B - y_A)^2}$$

**Q1. √âcrire une fonction `dist_carree(A,B)` qui prend en param√®tre deux tuples A et B contenant les coordonn√©es des points $A$ et $B$ et qui renvoie le carr√© de la distance AB (c'est-√†-dire: AB¬≤).**

````python
#Question 1
def dist_carree(A,B):
    '''Revoie le carr√© de la distance entre les deux coordon√©es de points donn√©es en param√®tre de la fonction (doit √™tre des tuples)'''
    return (B[0]-A[0])**2+(B[1]-A[1])**2
    
A=(1,2)
B=(3,4)
print(dist_carree(A,B))
````

## 2. Trouver le milieu

Les coordonn√©es du milieu de deux points $A(x_A,y_A)$ et $B(x_B,y_B)$ dont on conna√Æt les coordonn√©es dans un rep√®re orthonorm√© sont :

$$\left( \frac{x_A + x_B}{2} ,\frac{y_A + y_B}{2}\right)$$

**Q2.  √âcrire une fonction `milieu(A,B)` qui prend en param√®tre deux tuples A et B contenant les coordonn√©es des points $A$ et $B$ et qui renvoie les coordonn√©es de leur milieu sous forme d‚Äôun tuple.**

````python
# Question 2
def milieu(A,B):
    '''Revoie les coordon√©es du milieu des deux coordon√©es de points donn√©es en param√®tre de la fonction (doit √™tre des tuples)'''
    return ((A[0]+B[0])/2, (A[1]+B[1])/2)
    
A=(1,2)
B=(3,4)
print(milieu(A,B))
````

````python
# Question 3
def triplet_pythagoricien(n):
    '''Renvoie sous forme d'un tuple de tuples (tuple de triplets) tous les triplets Pythagoriciens d‚Äôentiers ‚â§ ùëõ, n √©tant le param√®tre de la fonction'''
    result = ()
    for a in range(n+1) :
        for b in range(a,n+1) :
            if ((a**2+b**2)**0.5 <= 100 and (a**2+b**2)**0.5-int((a**2+b**2)**0.5) == 0 and a != 0 and b != 0) :
                result += ((a,b,int((a**2+b**2)**0.5)),)
    return result

print(triplet_pythagoricien(100))
print(len(triplet_pythagoricien(100)))
````
